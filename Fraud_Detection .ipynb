Import the necessary **libraries**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

load the dataset

```python
credit_card_data=pd.read_csv(r'/content/creditcard.csv')
```

inspecting the first 5 data

```python
credit_card_data.head()
```

dataset information

```python
credit_card_data.info()
```

checking no of missing values in each columns

```python
credit_card_data.isnull().sum()
```

# distribution of *legit* transaction and fraudulent transactions

```python
credit_card_data['Class'].value_counts()
```

seperate the data for analysis

```python
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

# checking the shape of legit and fraud data

```python
print(legit.shape)
print(fraud.shape)
```

# compare the values for both transaction

```python

print(credit_card_data.groupby('Class').mean())
```

undersampling
# build a sample dataset containing similar distribution of normal transation and fraudlent transaction

```python
legit_sample=legit.sample(n=492)

#concatenate 2 dataframe
new_dataset=pd.concat([legit_sample,fraud],axis=0)
print(new_dataset.head())
```

```python
print(new_dataset['Class'].value_counts())
```

# splitting the data into features and traning

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
print(X)
print(Y)
```

# splitting the data into training set and testing dataset

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
print(X.shape,X_train.shape,X_test.shape)
```

# Model evaluation

```python
model=LogisticRegression()
model.fit(X_train,Y_train)
```

```python
print(f'accuracy of training data is {accuracy_score(model.predict(X_train),Y_train)}')
```

```python
print(f'accuracy of testing data is {accuracy_score(model.predict(X_test),Y_test)}')
```

make a prediction

```python
input_data=X_test.iloc[0]
print(input_data)
```

```python

if model.predict([input_data])[0]==0:
  print('legit transaction')
else:
  print('fraudlent transaction')
```
